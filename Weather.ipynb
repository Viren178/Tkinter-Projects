```python
from tkinter import *
import tkinter as tk
from geopy.geocoders import Nominatim
from tkinter import ttk,messagebox
from timezonefinder import TimezoneFinder
from datetime import datetime
import requests
import pytz

root=Tk()
root.title("Weather")
root.geometry("800x600+300+200")
root.resizable(False,False)

#functions

def getweather():
    city=search_bar.get()
    locater=Nominatim(user_agent="geoapiExercises")
    location=locater.geocode(city)
    tf=TimezoneFinder()
    result=tf.timezone_at(lng=location.longitude,lat=location.latitude)
    
    home=pytz.timezone(result)
    local_time=datetime.now(home)
    current_time=local_time.strftime("%I:%M %p")
    clock.config(text=current_time)
    name.config(text="CURRENT WEATHER")
    
    #api
    api="https://api.openweathermap.org/data/2.5/weather?q="+city+"&appid=efb3e2d0fc7fa12fb1ba79d78d399863"
    json_data= requests.get(api).json()
    condition=json_data["weather"][0]["main"]
    description=json_data["weather"][0]["description"]
    temp=int(json_data["main"]["temp"]-273.15)
    pressure=json_data["main"]["pressure"]
    humidity=json_data["main"]["humidity"]
    wind = json_data["wind"]["speed"]
    
    t.config(text=(temp,"°"))
    c.config(text=(condition,"|","feels" ,"like",temp,"°"))
    
    w.config(text=wind)
    d.config(text=description)
    h.config(text=humidity)
    p.config(text=pressure)

#Search
search_image=PhotoImage(file=r"C:\Users\A\Downloads\weather.png")
wimage=Label(image=search_image)
wimage.place(x=20,y=20)

search_bar=tk.Entry(root,justify="center", width=15,font=("poppins",24,"bold"),fg="white",bg="black")
search_bar.place(x=50,y=40)
search_bar.focus

#Search logo
search_icon=PhotoImage(file=r"C:\Users\A\Downloads\icons8-search-30.png")
image_icon=Button(image=search_icon,borderwidth=0,cursor="hand2",bg="#404040",command=getweather)
image_icon.place(x=290,y=44)

#time
name=Label(root,font=("ariel",17,"bold"))
name.place(x=60,y=100)
clock=Label(root,font=("Helvetica",17))
clock.place(x=60,y=150)

#label
label1=Label(root, text="Wind", font=("Helvetica",20,"bold"),bg="black",fg="white")
label1.place(x=65,y=500)
label2=Label(root, text="Pressure", font=("Helvetica",20,"bold"),bg="black",fg="white")
label2.place(x=205,y=500)
label3=Label(root, text="Humidity", font=("Helvetica",20,"bold"),bg="black",fg="white")
label3.place(x=405,y=500)
label4=Label(root, text="Discription", font=("Helvetica",20,"bold"),bg="black",fg="white")
label4.place(x=605,y=500)

t=Label(font=("ariel",70,"bold"),fg="#ee666d")
t.place(x=160,y=200)
c=Label(font=("ariel",15,"bold"))
c.place(x=160,y=300)


w=Label(text="...", font=("Helvetica",20,"bold"),bg="black",fg="white")
w.place(x=65,y=538)
p=Label(text="...", font=("Helvetica",20,"bold"),bg="black",fg="white")
p.place(x=205,y=538)
h=Label(text="...", font=("Helvetica",20,"bold"),bg="black",fg="white")
h.place(x=405,y=538)
d=Label(text="...", font=("Helvetica",20,"bold"),bg="black",fg="white")
d.place(x=605,y=538)




root.resizable(True,True)

root.mainloop()
```
